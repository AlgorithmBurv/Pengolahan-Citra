```python
# memanggil modul yang diperlukan
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

#bgr
img = cv2.imread('/content/foto.jpg')

#rgb
cat = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# tampilkan gambar awal tanpa filter
cv2_imshow( img)

# membuat filter: matriks berukuran 3x3
kernel = np.ones((3,3) ,np.float32)/9
print(kernel)

# lakukan filtering
hasil_filter = cv2.filter2D(img,-1,kernel)

cv2_imshow(hasil_filter)
```

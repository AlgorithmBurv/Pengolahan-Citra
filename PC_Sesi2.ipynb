```python
kendaraan = input("Kendaraan anda adalah (motor/mobil): ").lower()
bensin = input("Apakah bensin anda penuh (ya/tidak): ").lower()

# jika motor tidak ada bensin
if kendaraan == "motor" and bensin == "tidak":
    print("\n Silahkan naek ojek online")

# jika mobil tidak ada bensin
elif kendaraan == "mobil" and bensin == "tidak":
  print("\n Silahkan naek transportasi umum")

# jika motor dan mobil ada bensin
else:
  print("\n Silahkan berkendara, hati-hati....!")
```
